```python
from xml.dom import minidom
from dateutil.parser import parse
import re
```

```python
contestantDom = dom.getElementsByTagName('tr')
```

```python
problemsDom = contestantDom[0]
problems = [x.firstChild.data for x in problemsDom.getElementsByTagName('th')][2:-2]
```

```python
hailoDom = contestantsDom[0]
def parseDom(dom):
    return [x.firstChild.data for x in dom.getElementsByTagName('td')[1:-2]]
```

```python
scoreboard = map(parseDom, contestantsDom)
[s for s in scoreboard]
```

```python
handle = [x[0] for x in scoreboard]
handle
```

```python
def genSubmitions(substr):
    if substr == '-':
        return None
    else:
        sp = substr.split()
        last = sp[-1]
        times = int(re.findall('\d+', last)[0])
        datestr = ' '.join(sp[:-1])
        if not datestr:
            return None
        return (datestr, times)
```

```python
cured = [
    'hailo',
    'kvaleriano',
    'Robert_Alonso',
    'itu',
    'Rxcso',
    'jael',
    'kirasam',
    'Slsvcn',
    'Lucas97',
    'jafc',
    'juanjo12x',
    'Dan212P',
    'Gerard901',
    'The_Blitz',
]
len(cured)
```

```python
from contestant import *
from contest import *
submissions = []
for idy, row in enumerate(scoreboard):
    name = cured[idy]
    for idx, problemStr in enumerate(row[1:]):
        sub = genSubmitions(problemStr)
        if sub:
            date = sub[0]
            watimes = sub[1]
            for t in xrange(watimes):
               submissions.append(Submition(
                   handle = name,
                   veredict = Veredict.getVeredict('wrong'),
                   problem = problems[idx],
                   seconds = (parse(date) - parse('2014-09-11 19:07:00')).total_seconds() - 10
                )) 
            submissions.append(Submition(
               handle = name,
               veredict = Veredict.getVeredict('accepted'),
               problem = problems[idx],
               seconds = (parse(date) - parse('2014-09-11 19:07:00')).total_seconds()
            ))
```

```python
submissions.sort(key= lambda x: x.seconds)
```

```python
def splitEach(time_stream, seconds):
    end, start = seconds, 0
    ans, cur = [], []
    for idx, time in enumerate(time_stream): 
        if time.seconds > end:
            ans.append(cur)
            cur = [time]
            end += seconds
        else: 
            cur.append(time)
    if cur:
        ans.append(cur)
    return ans
```

```python
map(len, splitEach(submissions, 350))
```

```python
contest = Contest(cured, problems)
timer = 350
buckets = splitEach(submissions, timer)
scoreboards = []
for b in buckets:
    for s in b: 
        contest.apply(s)
    order = sorted(contest.getScoreboard(), key=lambda x: (-x.solved, x.penalty))
    scoreboards.append(order)
scoreboards
```

```python
f = open('../assets/data/7/scoreboard.json', 'w+')
import json
json_data = {
    'contestants': cured,
    'contests': [str(((idx + 1) * (timer/60)))+"'" for idx in  range(len(scoreboards))],
    'scoreboards': [[d.__dict__ for d in score] for score in scoreboards],
}
json.dump(json_data, f)
f.close()
```
